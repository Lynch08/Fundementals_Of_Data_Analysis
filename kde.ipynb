# PDF: Probability Density Function

```python
# Normal Distribution

def normalpdf(x, mu = 0.0, sigma = 1.0):
    """Calculates the PDF of a normal distribution"""
    numer = np.exp(-0.5 * np.square((x - mu) / sigma))
    denom = sigma * np.sqrt(2.0 * np.pi)
    return (numer / denom)
```

```python
# test value
normalpdf(0.0)
```

```python
# test values
normalpdf(np.array([-1.0, 0.0, 1.0]))
```

```python
# Some x Values
x = np.linspace(-5.0, 5.0 , 1000)

# Calculate our y values from x
y = normalpdf(x)

# Create a plot
fig, ax = plt.subplots()

# Plot the Values
ax.plot(x, y, 'r-')
```

# Fill between in plots

```python
# Some x Values
x = np.linspace(-5.0, 5.0 , 1000)

# Calculate our y values from x
y = normalpdf(x)

# Create a plot
fig, ax = plt.subplots()

# Plot the Values
ax.plot(x, y, 'k-')

# limit x values to -2.0 and 2.0
x = np.linspace(-2.0, 2.0 , 1000)

# Calculate the odf values
y = normalpdf(x)

# Fill between x axis and y - for for x values between -2.0 and 2.0
ax.fill_between(x, 0, y)

# Draw vertical lines
ax.axvline(-2.0, color = 'k', linestyle = ':')
ax.axvline(2.0, color = 'k', linestyle = ':')

# Draw some text
ax.text (0, 0.2, '~95%', horizontalalignment = 'center')
```

```python
# Probability of value less than -2.0
stats.norm().cdf(-2.0)
```

```python
# Probability of value less than 2.0
stats.norm().cdf(2.0)
```

```python
# Probability of value between -2.0 and 2.0
stats.norm().cdf(2.0) - stats.norm().cdf(-2.0)
```

```python
# Probability of value between -3.0 and 3.0
stats.norm().cdf(3.0) - stats.norm().cdf(-3.0)
```

```python
# Minimum and maximum values
low, high = sepdata.min(), sepdata.max()

#Create some space on the x axis
diff = high - low
low = low - diff
high = high + diff

# Some x Values
x = np.linspace(low, high , 1000)

# Create a plot
fig, ax = plt.subplots()

prev = 0 * x

for i in sepdata:
    # Calculate our y values from x
    y = normalpdf(x, mu=i, sigma = 0.1 )
    # Add the previous sums of the y values
    y = y + prev

    # Fill between x axis and y
    ax.fill_between(x, prev, y, alpha = 0.3)
    
    # Update the prev array
    prev = y

# Plot the Values
ax.plot(x, y, 'k-')
```
