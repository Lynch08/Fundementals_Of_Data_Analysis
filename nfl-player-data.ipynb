```python
# Read in the data
nfl_df = pd.read_csv('nfl.csv') 
```

```python
#take a peek at raw data
nfl_df.head()
```

```python
# I only want to look at Quarterbacks 
df = nfl_df.loc[nfl_df.loc[:, 'pos'] == 'QB']
```

```python
# These are the only columns I want
pos2 = df[['player', 'pass_att', 'pass_cmp','pass_yds', 'team' ]]
pos2.head()
```

```python
# This is weekly data
# So I want the attemts, completions and yards columns summed using the players name
pos3 = pos2.groupby('player')['pass_att', 'pass_cmp', 'pass_yds'].apply(lambda x : x.astype(int).sum())
pos3.head()
```

```python
# Resets the columns correctly after using groupby()
pos4 = pos3.reset_index()
```

```python
# I only want to see the players with over at least 750 pass completions
pos5 = pos4[pos4.pass_cmp >750]
```

```python
#Take a peek at your data
pos5
```

```python
# Sort the dataframe by pass completions
# I am only using pass completions as a filter - will not be represented in the graph plot
nfl_df = pos5
nfl_df = nfl_df.sort_values(by ='pass_att', ascending = False)
```

```python
# Peek at the data now I am ready to plot
nfl_df 
```

```python

fig, ax = plt.subplots(figsize=(20, 10))
x = nfl_df['pass_att'].values
y = nfl_df['pass_yds'].values
t = nfl_df['player'].values
ax.scatter(x, y)
ax.set_xlabel('Pass Att', fontname="Arial", fontsize=20)
ax.set_ylabel('Pass Yds', fontname="Arial", fontsize=20)
ax.set_title('Pass Att vs Pass Yds', fontname='Comic Sans MS',fontsize=50 )
ax.set_facecolor('y')
for i, txt in enumerate(t):
    ax.annotate(txt, xy=(x[i], y[i]), xytext=(x[i], y[i]+0.5), ha='center', weight = 'bold')
for i, txt in enumerate(t):
    ax.annotate(txt, xy=(x[i], y[i]), xytext=(x[i], y[i]+0.5), ha='center', weight = 'bold')
plt.show()
```
