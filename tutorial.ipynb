```python
# efficent numerical arreys
import numpy as np
#plotting
import matplotlib.pyplot as plt
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

# Functional Plot

```python
x_1 = np.linspace(0,7,10)
y_1 = x_1**2
plt.plot(x_1, y_1)
plt.title('Days squared chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
plt.show()
```

# Print Mutiple plots

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1, 'g')
plt.subplot(1,2,2)
plt.plot(x_1,y_1, 'r')
```

# Using figure Objects
A figure is an object that is going to contain all the plot outlets and axis'.


```python
# Figuresize and dots/pixals per inch
fig_1 = plt.figure(figsize=(8,6), dpi = 100)

axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
# Labels, title, legend
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1, y_1, label  = 'x / x^2')
axes_1.plot(y_1, x_1, label  = 'x^2 / x')
# legend loc = 0 allows pyplot to try to figure out the best place to place the legend
# loc 1 = upper right, 2 = upper left, 3 = lower left, 4 = lower right
axes_1.legend(loc=0)

#adding a ploy inside a plot
axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1, y_1, 'r')

# adding text to explain the plot
axes_2.text(1,70,'Explanation of data')
```

# Subplots

```python
# number of plots
fig_2, axes_2 = plt.subplots(figsize = (8,4), nrows = 1, ncols = 3)

#puts space between plots
plt.tight_layout()

#access plots using index - 0 index so [1] will allow me to manipulate the second plot 
axes_2[0].set_title('Plot 1')
axes_2[1].set_title('Plot 2')
axes_2[2].set_title('Plot 3')

axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel = ('X Squared')
axes_2[1].plot(x_1, y_1)
```

# Appearance options
Default colors (b: blue, g: green, r: red, c: cyan, m: magenta,y: yellow, k: black, w: white)color="0.75" creates a 75% gray
You can use hexcodes color="#eeefff"
You can use color names found next like this color="burlywood"
https://en.wikipedia.org/wiki/Web_colors
alpha defines the percentage of opacity

The default line width is 1, so to double it put in 2 and so forth

There are many line styles 
matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html
You can also provide a sample like '-.'

Markers can mark your provided points on the graph
https://matplotlib.org/3.3.0/api/markers_api.html
You can change the markersize as well

markerfacecolor changes the marker fill color
markeredgecolor changes the marker stroke color
markeredgewidth changes the markers stroke size

```python
#brightening up the plot
fig_3 = plt.figure(figsize = (6,4))
axes_3 = fig_3.add_axes([0,0,1,1])
axes_3.plot(x_1,y_1, color='cyan', alpha = .90, lw = 2, ls = '-.', marker = 'o', markersize = 7, markerfacecolor = 'b')

#set limits on axes
axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

# brighten up background - or make it so ugly it draws attention
axes_3.grid(True, color = 'g', dashes =(5,2,1,2) )
axes_3.set_facecolor(color = 'y')
```

# Working with Pandas

```python
nfl_df = pd.read_csv('nfl2.csv') 
nfl_df = nfl_df.sort_values(by ='pass_att', ascending = False)

fig, ax = plt.subplots(figsize=(15, 10))
x = nfl_df['pass_att'].values
y = nfl_df['pass_cmp'].values
t = nfl_df['player'].values
ax.scatter(x, y)
ax.set_xlabel('Pass Att')
ax.set_ylabel('Pass Cmp')
ax.set_title('Pass Att vs Pass Cmp')
for i, txt in enumerate(t):
    ax.annotate(txt, xy=(x[i], y[i]), xytext=(x[i], y[i]+0.5), ha='center')
plt.show()

```
